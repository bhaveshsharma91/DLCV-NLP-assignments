```python
import urllib.request
from bs4 import BeautifulSoup

resp=urllib.request.urlopen('http://vincentarelbundock.github.io/Rdatasets/datasets.html')
html=resp.read()

# Parse the html file
soup = BeautifulSoup(html, 'html.parser')

# Format the parsed html file
strhtm = soup.prettify()

# Print the first few characters
```

```python
soup.prettify()[0:300]
```

```python
table19 = soup.findAll("td",{"class":"cellinside"})
table19
```

```python
import csv
with open("data.csv", "wt+", newline="") as f:
    writer = csv.writer(f)
    count=0
    rssd=[]
    for i in range(len(table19)//12):
        count+=1
        rssd=[]
        i19=table19[12*i:10*(i+1)]
        for j in i19:
            j = str(j)
            j = j.replace("     ",'')
            j = j.replace("   ","")
            j = j.replace("  ","")

            rssd.append(j[23:-6])
        writer.writerow(rssd)

```
